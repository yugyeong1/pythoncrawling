```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

url = 'https://www.hankyung.com/economy?page=1'
res = requests.get(url)
```

```python
soup = bs(res.text, 'lxml')
ul = soup.find("ul",{"class":"news-list"}).findAll("li")
```

```python
news = []
for li in ul:
    data = li.find("div",{"class":"txt-cont"})
    news.append({
    'title': data.find("a",{"data-pm":"N"}),
    'text': data.find("p", {"class":"lead"})
    })
```

```python
print(news)
```
